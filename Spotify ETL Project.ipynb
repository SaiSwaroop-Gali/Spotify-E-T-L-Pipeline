```python
album_details = []
for row in data ['items']:
    album_name = row['track']['album']['name']
    album_id = row['track']['album']['id']
    release_date = row['track']['album']['release_date']
    album_url=row['track']['album']['external_urls']['spotify']
    total_tracks= row['track']['album']['total_tracks']
    album_items={'Album name':album_name, 'Id':album_id,'Releasedate':release_date, 'URL':album_url
                  , 'No.of tracks':total_tracks}
    album_details.append(album_items)
```

```python
album_df= pd.DataFrame.from_dict(album_details) #it can also be written as ***album_df= pd.DataFrame(album_details)***
album_df.head()
```

```python
album_df['Releasedate']=pd.to_datetime(album_df['Releasedate'],format='mixed', dayfirst=True)
album_df.head()
```

```python
artist_list = []                         #dont confuse idiot #for calling dictionary data we need to provide (key: values/key,values) even for loops
for row in data['items']:
    for key, value in row.items():
        if key == 'track':
            for artist in value['artists']:
                artist_data={'artist_id':artist['id'],'artist_name':artist['name'],'artist_url':artist['href']}
                artist_list.append(artist_data)
```

```python
artist_df=pd.DataFrame.from_dict(artist_list)
artist_df.head()
```

```python
song_list = []
for row in data['items']:
    song_id = row['track']['id']
    song_name = row['track']['name']
    song_duration = row['track']['duration_ms']
    song_url = row['track']['external_urls']['spotify']
    song_popularity = row['track']['popularity']
    song_added = row['added_at']
    album_id = row['track']['album']['id']
    artist_id = row['track']['album']['artists'][0]['id']
    song_element = {'song_id':song_id,'song_name':song_name,'duration_ms':song_duration,'url':song_url,
                    'popularity':song_popularity,'song_added':song_added,'album_id':album_id,
                    'artist_id':artist_id
                   }
    song_list.append(song_element)
```

```python
song_list
```

```python
song_df=pd.DataFrame.from_dict(song_list)
```
